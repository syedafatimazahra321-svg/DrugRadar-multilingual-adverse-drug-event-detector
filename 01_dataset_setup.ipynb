splits it into train/test, and saves it to `data/raw/` folder.

## Cell 2 — Load the dataset from HuggingFace

```python
from datasets import load_dataset
import pandas as pd
from collections import Counter

print("Loading ADE Corpus V2 from HuggingFace...")
ds = load_dataset("ade-benchmark-corpus/ade_corpus_v2", "Ade_corpus_v2_classification")
df_all = ds["train"].to_pandas()

print(f"Loaded {len(df_all)} rows")
print(f"Columns : {list(df_all.columns)}")
print(f"\nLabel breakdown:")
counts = Counter(df_all["label"])
print(f"  0 = No ADE : {counts[0]} rows ({counts[0]/len(df_all)*100:.1f}%)")
print(f"  1 = ADE    : {counts[1]} rows ({counts[1]/len(df_all)*100:.1f}%)")
print(f"\nSample rows:")
print(df_all.head(5).to_string())
```

## Cell 3 — Split into train and test sets

```python
from sklearn.model_selection import train_test_split

# 80% train 20% test using same random seed every time so results are reproducible
train_df, test_df = train_test_split(
    df_all,
    test_size=0.20,
    random_state=42,
    stratify=df_all["label"]   # to keep same ADE ratio in both splits
)

print(f"Train rows : {len(train_df)}")
print(f"Test rows  : {len(test_df)}")
print(f"\nTrain ADE rate : {train_df['label'].mean()*100:.1f}%")
print(f"Test  ADE rate : {test_df['label'].mean()*100:.1f}%")
```

## Cell 4 — Save to CSV files

```python
train_df.to_csv("/content/drive/MyDrive/DrugRadar/data/raw/train.csv", index=False)
test_df.to_csv("/content/drive/MyDrive/DrugRadar/data/raw/test.csv", index=False)
print("Files saved to Google Drive!")

print("Saved data/raw/train.csv")
print("Saved data/raw/test.csv")
print("\nVerifying files exist...")

import os
for f in ["/content/drive/MyDrive/DrugRadar/data/raw/train.csv", "/content/drive/MyDrive/DrugRadar/data/raw/test.csv"]:
    size_kb = os.path.getsize(f) // 1024
    print(f"  {f}  ({size_kb} KB)")

print("\n Dataset setup complete!")
```
